```python
df.plot()  #since its a seasonal data,we use dickyfuller test
```

```python
from statsmodels.tsa.stattools import adfuller
```

```python
def adf_test(series):
    result=adfuller(series)
    print('ADF statistics: {}'.format(result[0]))
    print('P-value: {}'.format(result[1]))
    
    if result[1]<=0.05:
        print('reject null hypothesis')
        
    else:
        print('accept null hypothesis')
```

```python
adf_test(df['Thousands of Passengers'])
```

```python
#sice it is not stateoney we do differencing   

df['first_difference']=df['Thousands of Passengers']-df['Thousands of Passengers'].shift(1)   #shift(1) will move one step back
```

```python
df
```

```python
df.dropna(inplace=True)
```

```python
adf_test(df['first_difference'])  #now also its not stationary data,hence we do second difference
```

```python
df['second_difference']=df['first_difference']-df['first_difference'].shift(1) 
```

```python
df.head()
```

```python
df.dropna(inplace=True)
```

```python
adf_test(df['second_difference'])   #now it becomes stationary data
```

```python
#since it is a seasonal data we have 12 months hence we have to make a 12 diffence data
df['12_difference']=df['Thousands of Passengers']-df['Thousands of Passengers'].shift(12) 
```

```python
adf_test(df['12_difference'].dropna())  
```

```python
from datetime import datetime,timedelta
```

```python
train_data_end=datetime(1955,12,1)
test_data_end=datetime(1960,12,1)
```

```python
train_data=df[:train_data_end]
test_data=df[train_data_end+timedelta(days=1):test_data_end]
```

```python
pred_start_date=test_data.index[0]
pred_end_date=test_data.index[-1]
```

```python
from statsmodels.tsa.arima_model import ARIMA
```

```python
train_data.head()
```

```python
import statsmodels.api as sm
```

```python
model_arima=sm.tsa.arima.ARIMA(train_data['Thousands of Passengers'],order=(14,2,0))   #order=p value,d value(difference),q value
```

```python
model_arima_fit=model_arima.fit()
```

```python
model_arima_fit.summary()
```

```python
pred=model_arima_fit.predict(start=pred_start_date,end=pred_end_date)
residual=test_data['Thousands of Passengers']-pred
```

```python
residual
```

```python
model_arima_fit.resid.plot(kind='kde')
```

```python
test_data['predicted_arima']=pred
```

```python
test_data[['Thousands of Passengers','predicted_arima']].plot()
```

```python
model_sarima=sm.tsa.statespace.SARIMAX(train_data['Thousands of Passengers'],order=(3,0,5),seasonal_order=(1,1,1,12)) 
```

```python
model_sarima_fit=model_sarima.fit()
```
